```python
# Hypothesis check stionanry data
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['volume'])
```

```python
def adf_check(time_series):
    
    result = adfuller(time_series)
    print('Agumented Dicky-Fuller Test')
    labels = ['ADF Test Statistic','p-value','# of legs','Num of oberservations used']
    
    for value,label in zip(result, labels):
        print(label+ " : "+str(value))
        
    if result[1] <= 0.05:
        print("strong evidence against null hypothesis")
        print('reject null hypothesis')
        print('Data has no unit root and it is stationary')
    else:
        print('Weak evidence against null hypothesis')
        print('Fail to reject null hypothesis')
        print('Data has a unit root, it is non-stationary')
```

```python
adf_check(df['volume'])
```

```python
adf_check(df['close'])
```

```python
adf_check(df['open'])
```
